```python
import requests as req
from bs4 import BeautifulSoup as bs
import csv
import os
import pandas as pd
```

```python
HEADER = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36'
FILENAME = 'kontan.csv'

def load_existing_links(filename):
    existing_links = set()
    # Jika file sudah ada, baca file dan ambil link yang sudah ada
    if os.path.exists(filename):
        with open(filename, 'r', newline='', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            for row in reader:
                existing_links.add(row['link'])
    return existing_links

def kontan():
    existing_links = load_existing_links(FILENAME)
    a = 1
    # Jika file belum ada, tulis header kolom. Jika sudah ada, tidak menimpa file.
    if not os.path.exists(FILENAME):
        with open(FILENAME, 'w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow(['judul', 'tanggal', 'link', 'paragraf', 'media'])

    for page in range(20, 500):
        try:
            url = f'https://www.kontan.co.id/search/?search=tarif-trump&per_page={page}'
            r = req.get(url, headers={"User-Agent": HEADER})
            soup = bs(r.text, 'lxml')
            list_berita = soup.find('div', class_='list-berita')

            if list_berita:
                berita = list_berita.find_all('div', class_='sp-hl linkto-black')

                if not berita:
                    print(f"Tidak ada berita ditemukan pada halaman {page}")
                    continue

                for item in berita:
                    try:
                        link_element = item.find('a')
                        if not link_element:
                            continue

                        link = link_element['href']
                        # Add 'https:' to the beginning of the link if it's missing
                        if not link.startswith('http'):
                            link = 'https:' + link if link.startswith('//') else 'https://' + link

                        # Skip jika link sudah ada pada file CSV sebelumnya
                        if link in existing_links:
                            continue

                        # Ambil judul dari halaman daftar
                        judul = item.text.strip()

                        # Request ke halaman artikel individual
                        article_r = req.get(link, headers={"User-Agent": HEADER})
                        article_soup = bs(article_r.text, 'lxml')

                        # Ambil tanggal dari halaman artikel
                        tanggal_element = article_soup.find('div', class_='fs14 ff-opensans font-gray')
                        tanggal = tanggal_element.text.strip() if tanggal_element else "Tanggal tidak ditemukan"

                        # Ambil konten artikel
                        isi = article_soup.find('div', class_='tmpt-desk-kon')
                        if isi:
                            paragraphs = [p.get_text(separator=" ", strip=True) for p in isi.find_all('p')]
                            paragraf = " ".join(paragraphs)
                        else:
                            paragraf = "Konten tidak ditemukan"

                        media = "Kontan"

                        print(f'done[{a}] -> {judul} {tanggal}')
                        a += 1

                        # Tulis data berita ke file CSV
                        with open(FILENAME, 'a', newline='', encoding='utf-8') as file:
                            writer = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                            writer.writerow([judul, tanggal, link, paragraf, media])

                        # Tambahkan link ke set agar tidak terjadi duplikasi selama satu run
                        existing_links.add(link)

                        # Delay untuk menghindari rate limiting
                        time.sleep(random.uniform(1, 3))

                    except Exception as e:
                        print(f"Error saat mengolah artikel: {e}")
                        continue
            else:
                print(f"Tidak ada berita ditemukan pada halaman {page}")

            # Delay antara halaman
            time.sleep(random.uniform(2, 5))

        except Exception as e:
            print(f"Error pada halaman {page}: {e}")
            continue

if __name__ == "__main__":
    kontan()
```

```python
import pandas as pd

# 1. Baca dataset
df = pd.read_csv('kontan.csv')

# 2. Fungsi klasifikasi berdasarkan isi artikel
def klasifikasi_berita(teks):
    word_count = len(str(teks).split())
    # paragraph_count = str(teks).count('\n') + 1
    if word_count < 400:
        return "Straight News"
    else:
        return "In-Depth News"

# 3. Terapkan ke dataset
df['jenis'] = df['paragraf'].apply(klasifikasi_berita)

# 4. Simpan ke CSV baru
df.to_csv('kontan_labeled.csv', index=False, encoding='utf-8')
print("Berhasil diklasifikasikan dan disimpan ke 'kontan_labeled.csv'")
```
